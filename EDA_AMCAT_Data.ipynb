## Exploratory Data Analysis

```python
# Salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=30, kde=True, color='skyblue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# College GPA distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['collegeGPA'], bins=30, kde=True, color='salmon')
plt.title('College GPA Distribution')
plt.xlabel('College GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
# Gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Degree distribution
plt.figure(figsize=(12, 6))
sns.countplot(x='Degree', data=df, palette='Set3')
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot: Salary vs. College GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df, color='skyblue')
plt.title('Salary vs. College GPA')
plt.xlabel('College GPA')
plt.ylabel('Salary')
plt.show()
```

```python
# Box plot: Salary vs. Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Salary', data=df, palette='Set2')
plt.title('Salary vs. Gender')
plt.xlabel('Gender')
plt.ylabel('Salary')
plt.show()
```

###1. What are the significant factors influencing salary levels among job seekers?

```python
# Correlation heatmap to identify significant factors
plt.figure(figsize=(20, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Variables')
plt.show()
```

###2. Are there any correlations between educational qualifications, specialization, and salary levels?



```python
# Box plot: Salary vs. Degree
plt.figure(figsize=(12, 6))
sns.boxplot(x='Degree', y='Salary', data=df, palette='Set3')
plt.title('Salary vs. Degree')
plt.xlabel('Degree')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(25, 10))
sns.barplot(x='Specialization', y='Salary', data=df, ci=None)  # ci=None to remove error bars
plt.title('Average Salary Levels by Specialization')
plt.xlabel('Specialization')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

###3. How do personal characteristics such as gender, age, and location impact salary expectations?

```python
# Salary levels by gender (bar plot)
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Salary', data=df, ci=None)
plt.title('Average Salary Levels by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')
plt.show()

# Salary levels by JobCity (bar plot)
plt.figure(figsize=(20,10))
sns.barplot(x='JobCity', y='Salary', data=df, ci=None)
plt.title('Average Salary Levels by JobCity')
plt.xlabel('JobCity')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
# Grouping the data by 'JobCity' and calculating the mean salary for each city
avg_salary_by_city = df.groupby('JobCity')['Salary'].mean().reset_index()

# Finding the city with the highest average salary
highest_avg_salary_city = avg_salary_by_city.loc[avg_salary_by_city['Salary'].idxmax()]

print("City with the highest average salary:")
print(highest_avg_salary_city)
```

###4. Can we identify any trends or patterns in salary distributions across different job designations and cities?

```python
# Salary levels by Designation (bar plot)
plt.figure(figsize=(20, 10))
sns.barplot(x='Salary', y='Designation', data=df, ci=None)
plt.title('Average Salary Levels by Designation')
plt.xlabel('Average Salary')
plt.ylabel('Designation')
plt.xticks(rotation=90)
plt.show()
```

```python
# Grouping the data by 'Designation' and calculating the mean salary for each level
avg_salary_by_designation = df.groupby('Designation')['Salary'].mean().reset_index()

# Finding the level with the highest average salary
highest_avg_salary_level = avg_salary_by_designation.loc[avg_salary_by_designation['Salary'].idxmax()]

print("Level with the highest average salary:")
print(highest_avg_salary_level)
```
